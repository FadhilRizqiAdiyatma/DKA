```python
import networkx as nx              # import library networkx untuk membuat graf
import matplotlib.pyplot as plt   # import matplotlib untuk visualisasi

# membuat graf kosong (tidak berarah / undirected graph)
G = nx.Graph()                            

# menambahkan edge (jalur antar kota) beserta bobot (jarak)
G.add_weighted_edges_from([               
("Arad","Zerind",75),          # Arad ke Zerind jaraknya 75
("Zerind","Oradea",71),
("Oradea","Sibiu",151),
("Arad","Sibiu",140),
("Arad","Timisoara",118),
("Timisoara","Lugoj",111),
("Lugoj","Mehadia",70),
("Mehadia","Drobeta",75),
("Drobeta","Craiova",120),
("Craiova","Pitesti",138),
("Craiova","Rimnicu Vilcea",146),
("Sibiu","Rimnicu Vilcea",80),
("Sibiu","Fagaras",99),
("Rimnicu Vilcea","Pitesti",97),
("Fagaras","Bucharest",211),
("Pitesti","Bucharest",101),
("Bucharest","Giurgiu",90),
("Bucharest","Urziceni",85),
("Urziceni","Hirsova",98),
("Hirsova","Eforie",86),
("Urziceni","Vaslui",142),
("Vaslui","Iasi",92),
("Iasi","Neamt",87)
])

# menentukan posisi setiap node (kota) agar layout graf rapi dan sesuai peta
pos = {                                  
"Arad":(-6,0),
"Zerind":(-5,2),
"Oradea":(-3,3),
"Sibiu":(-2,0),
"Fagaras":(0,0),
"Rimnicu Vilcea":(-1,-1),
"Pitesti":(0,-2),
"Craiova":(-2,-3),
"Drobeta":(-4,-3),
"Mehadia":(-5,-2),
"Lugoj":(-5,-1),
"Timisoara":(-6,-1),
"Bucharest":(2,-2),
"Giurgiu":(2,-4),
"Urziceni":(4,-2),
"Hirsova":(6,-2),
"Eforie":(7,-3),
"Vaslui":(5,0),
"Iasi":(4,1),
"Neamt":(3,2)
}

# mengambil atribut weight (jarak) dari setiap edge untuk ditampilkan sebagai label
labels = nx.get_edge_attributes(G,'weight')   

# mengatur ukuran gambar (figure)
plt.figure(figsize=(12,7))                   

# menggambar graf (node dan edge)
nx.draw(G,pos,                                
        with_labels=True,        # menampilkan nama kota
        node_size=2000,          # ukuran node (lingkaran)
        node_color="white",      # warna isi node
        edgecolors="black",      # warna garis tepi node
        font_size=9)             # ukuran teks label kota

# menampilkan label bobot (jarak) pada setiap edge
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)  

# memberi judul pada graf
plt.title("Graf Peta Kota-kota Besar di Eropa")         

# menampilkan hasil visualisasi
plt.show()
```

```python
import networkx as nx              # import library networkx untuk membuat dan mengelola graf
import matplotlib.pyplot as plt   # import matplotlib untuk menampilkan visualisasi graf

# membuat graf kosong (undirected graph / graf tidak berarah)
G = nx.Graph()                            

# menambahkan edge (hubungan antar kota) beserta bobot (jarak dalam km)
G.add_weighted_edges_from([               
("Jakarta","Cirebon",327),      # jarak Jakarta ke Cirebon = 327 km              
("Jakarta","Bandung",270),      # Jakarta ke Bandung = 270 km
("Bandung","Cirebon",120),      # Bandung ke Cirebon = 120 km              
("Bandung","Yogyakarta",373),   # Bandung ke Yogyakarta = 373 km         
("Cirebon","Semarang",305),     # Cirebon ke Semarang = 305 km        
("Cirebon","Yogyakarta",210),   # Cirebon ke Yogyakarta = 210 km            
("Semarang","Yogyakarta",109),  # Semarang ke Yogyakarta = 109 km          
("Semarang","Surakarta",97),    # Semarang ke Surakarta = 97 km           
("Yogyakarta","Surakarta",60),  # Yogyakarta ke Surakarta = 60 km        
("Semarang","Surabaya",369),    # Semarang ke Surabaya = 369 km
("Yogyakarta","Malang",370),    # Yogyakarta ke Malang = 370 km
("Surabaya","Malang",94)        # Surabaya ke Malang = 94 km
])

# menentukan posisi setiap node (kota) agar tampilan graf menyerupai peta Jawa
pos = {
"Jakarta":(-6,2),     # posisi Jakarta
"Bandung":(-5,0),     
"Cirebon":(-3,2),
"Semarang":(-1,2),
"Yogyakarta":(-2,0),
"Surakarta":(0,0),
"Surabaya":(2,2),
"Malang":(2,0)
}

# mengambil data bobot (weight) dari setiap edge untuk ditampilkan sebagai label
labels = nx.get_edge_attributes(G,'weight')  

# mengatur ukuran tampilan gambar (figure)
plt.figure(figsize=(10,5))                    

# menggambar graf (node dan edge)
nx.draw(G,pos,                                
        with_labels=True,     # menampilkan nama kota pada node                    
        node_size=2000,       # ukuran lingkaran node                 
        node_color="white",  # warna isi node                 
        edgecolors="black",  # warna garis tepi node              
        font_size=10)        # ukuran tulisan nama kota

# menampilkan label bobot (jarak) pada setiap edge
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels) 

# memberi judul pada graf
plt.title("Graf Kota-kota Besar di Pulau Jawa")       

# menampilkan hasil graf ke layar
plt.show()
```
